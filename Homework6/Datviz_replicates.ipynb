# Data Visualization of gene variant transcriptions groups

```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('serialdat.csv')
#str(df["10^x copies"])
# Display the first 5 rows of the DataFrame
print(df.head())
```

```python
df['10^x copies'] = df['10^x copies'].astype(str)
```

```python
df = df.dropna()
```

```python
df['SUMOvar'] = df['SUMOvar'].astype(str).apply(lambda x: x.split('-')[0])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# List of groups to plot
replicate =['6.0', '5.0', '4.0', '3.0', '2.0', '1.0']
fig, ax = plt.subplots(figsize=(10, 8))
# Iterate through the groups
for i in replicate:
    # Subset to the airline
    subset = df[df['10^x copies'] == i]
    
    # Draw the density plot
    sns.distplot(subset['Average Cq'], hist = False, kde = True,
                 kde_kws = {'linewidth': 2},
                 label = i)

# Plot formatting
plt.legend(prop={'size': 8}, title = 'Gene variant groups')
plt.title('Density Plot of gene variant transcriptions')
plt.xlabel('Average replications of the variant transcriptions')
plt.ylabel('Density')
```

From the above plots of the groups, As you move from Group 6 to group 1, you will notice the desity of the average replication value increase as the replication value increase.And I noticed an overlap in the second peaks of group 6 with the first peak of group 5, same with group 5 and 4, 4 and 3, 3 and 2  but 2 and 1 have just one peak.

```python
plt.figure(figsize=(8,6))
# create boxplots of the three rplicates using seaborn
sns.boxplot(data=df[["Replicate 1","Replicate 2","Replicate 3"]])


# set title and axis labels
plt.title("Boxplots of Three Replicates")
plt.ylabel("Values")

# show the plot
plt.show()
plt.show()
```

From the above plot, there seems to be no significat difference between the three replicates. Thogh replicate one seems to be slightly different from the replicate 2 and 3.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



sns.set_style("whitegrid")
sns.set_palette("Set2")

fig, ax = plt.subplots(figsize=(10, 8))

sns.boxplot(x='10^x copies', y='Average Cq', data=df,ax=ax)
sns.swarmplot(x='10^x copies', y='Average Cq', data=df, color='black',ax=ax)

ax.set_xlabel("Groups",fontsize=14)
ax.set_ylabel("Values",fontsize=14)
ax.set_title("Distribution of Values for Different Groups",fontsize=16)

plt.show()
```
